## Data Visualization with Matplotlib

Matplotlib is a powerful plotting library for Python that enables the creation of static, animated, and interactive visualizations. It is widely used for data visualization in data science and analytics.

In this lesson, we will cover the basics of Matplotlib, including creating various types of plots and customizing them.

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [2,3,4,5,6]

plt.plot(x,y)
```

```python
x = [1,2,3,4,5]
y = [2,3,4,5,6]

plt.plot(x,y)
plt.xlabel('X Axis')
plt.ylabel('Y axis')
plt.title("Example of Matplot Lib")
plt.show()
```
