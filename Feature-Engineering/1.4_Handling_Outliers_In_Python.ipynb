## 5 number Summanry And Box Plot

```python
# Minimum, Maximum, Median, Q1, Q2, IQR
```

```python
import numpy as np
```

```python
lst_marks = [42, 43, 78, 56, 82, 66, 72, 5, 23, 33, 47, 68, 99]

minimum, Q1, median, Q3, maximum = np.quantile(lst_marks,[0, 0.25,0.50,0.75,1.0])
```

```python
print(minimum, Q1, median, Q3, maximum)
```

```python
IQR = Q3-Q1
print(IQR)
```

```python
lower_fence = Q1-1.5*(IQR)
higher_fence = Q3+1.5*(IQR)
```

```python
lower_fence
```

```python
higher_fence
```
